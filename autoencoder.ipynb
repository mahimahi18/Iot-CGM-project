```python
# ---------- PASS 1: Chunk â†’ Fault Injection â†’ Clean â†’ Scale â†’ Save Preprocessed ----------
import os
import numpy as np
import pandas as pd
import random
from sklearn.preprocessing import StandardScaler

from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/iot/master_cgm_clean_no_faults.csv'
preprocessed_path = "/content/cgm_preprocessed.csv"
chunk_size = 50000

# Fit scaler only once using first 100k rows
print("ðŸ“¥ Fitting scaler on first 100k rows...")
sample = pd.read_csv(file_path, nrows=100000)
sample = sample.dropna(subset=["glucose"])
scaler = StandardScaler().fit(sample[["glucose"]].astype(float))

# Remove old preprocessed file
if os.path.exists(preprocessed_path):
    os.remove(preprocessed_path)

reader = pd.read_csv(file_path, chunksize=chunk_size)
first = True
chunk_id = 0

for chunk in reader:
    chunk_id += 1
    print(f"ðŸ”¸ Pass 1: Processing chunk {chunk_id}...")

    chunk = chunk.dropna(subset=["glucose"]).copy()
    chunk["glucose"] = pd.to_numeric(chunk["glucose"], errors="coerce")
    chunk = chunk.dropna(subset=["glucose"])

    chunk["glucose"] = chunk["glucose"].fillna(method="ffill")
    chunk["glucose_scaled"] = scaler.transform(chunk[["glucose"]]).astype(np.float32)

    # Save preprocessed file
    chunk.to_csv(
        preprocessed_path,
        mode="w" if first else "a",
        index=False,
        header=first
    )
    first = False

print("\nâœ… PASS 1 COMPLETE â†’ Preprocessed file saved at:")
print(preprocessed_path)
```

```python
# ---------- PASS 2: Train Autoencoder on Preprocessed Data ----------
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import regularizers
import numpy as np
import pandas as pd

print("\nðŸ§  PASS 2: Loading training sample...")

sample = pd.read_csv(preprocessed_path, usecols=["glucose_scaled"], nrows=120000)
X = sample[["glucose_scaled"]].values.astype(np.float32)

X_train, X_val = train_test_split(X, test_size=0.2, random_state=42)

# Build autoencoder
input_layer = Input(shape=(1,))
encoder = Dense(1, activation="relu", activity_regularizer=regularizers.l1(1e-6))(input_layer)
decoder = Dense(1, activation="linear")(encoder)
autoencoder = Model(input_layer, decoder)

autoencoder.compile(optimizer="adam", loss="mse")

print("\nðŸ§  Training autoencoder...")
autoencoder.fit(X_train, X_train,
                epochs=30,
                batch_size=64,
                validation_data=(X_val, X_val),
                verbose=1)

# Determine threshold
val_pred = autoencoder.predict(X_val, verbose=0)
val_err = np.abs(X_val - val_pred).flatten()
threshold = float(np.mean(val_err) + 2.5 * np.std(val_err))

print(f"\nðŸ”¹ Autoencoder Anomaly Threshold: {threshold:.6f}")
```

```python
# ---------- PASS 3: Detect anomalies using trained autoencoder ----------
output_path = '/content/drive/MyDrive/iot/autoencoder_res.csv'

if os.path.exists(output_path):
    os.remove(output_path)

reader = pd.read_csv(preprocessed_path, chunksize=50000)
first = True
cid = 0

for chunk in reader:
    cid += 1
    print(f"ðŸ”¸ PASS 3: Scoring chunk {cid}...")
    original = chunk[["glucose_scaled"]].values.astype(np.float32)
    Xc = original.copy()

    n_faults = int(0.03 * len(chunk))  # 3% faults
    fault_idx = np.random.choice(chunk.index, size=n_faults, replace=False)

    chunk["sensor_fault"] = 0   # ground truth indicator

    for i in fault_idx:
        f_type = random.choice(["spike", "dropout", "drift", "flatline"])

        try:
            if f_type == "spike":
                Xc[i] *= np.random.uniform(1.5, 3.0)

            elif f_type == "dropout":
                Xc[i] = np.nan  # missing value

            elif f_type == "drift":
                Xc[i] += Xc[i] * np.random.uniform(-0.4, 0.4)

            elif f_type == "flatline":
                if i + 3 < len(Xc):
                    val = Xc[i]
                    Xc[i:i+3] = val  # constant region

            chunk.loc[i, "sensor_fault"] = 1
        except:
            pass
    # Predict anomalies
    #Xc = chunk[["glucose_scaled"]].values.astype(np.float32)
    recon = autoencoder.predict(Xc, verbose=0).flatten()
    err = np.abs(Xc.flatten() - recon)

    chunk["recon"] = recon
    chunk["recon_error"] = err
    chunk["is_anomaly"] = err > threshold

    # Include predicted_label column for classifier block
    chunk["predicted_label"] = pd.Series(["N/A"] * len(chunk), dtype="string")

    chunk.to_csv(
        output_path,
        mode="w" if first else "a",
        index=False,
        header=first
    )
    first = False

print("\nâœ… PASS 3 COMPLETE â†’ Final anomaly output saved at:")
print(output_path)
```

```python
# ---------- AUTOENCODER PERFORMANCE METRICS ----------
import pandas as pd
from sklearn.metrics import confusion_matrix

print("\nðŸ“Š Evaluating Autoencoder anomaly detection performance...")

# Load the PASS 3 full output
res = pd.read_csv(output_path)

# Ground truth faults inserted earlier
y_true = res["sensor_fault"].astype(int)

# Predicted anomaly flag from reconstruction error threshold
y_pred = res["is_anomaly"].astype(int)

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

if cm.shape == (2, 2):
    tn, fp, fn, tp = cm.ravel()
else:
    raise RuntimeError("Confusion matrix not 2x2 â€” check label formatting")

# Metrics
total = tp + tn + fp + fn + 1e-9
acc_if  = (tp + tn) / total
prec_if = tp / (tp + fp + 1e-9)
rec_if  = tp / (tp + fn + 1e-9)
f1_if   = 2 * prec_if * rec_if / (prec_if + rec_if + 1e-9)

print("\n=== Autoencoder (Unsupervised) Metrics ===")
print(f"Accuracy:  {acc_if:.3f}")
print(f"Precision: {prec_if:.3f}")
print(f"Recall:    {rec_if:.3f}")
print(f"F1-score:  {f1_if:.3f}")

print("\nConfusion Matrix [tn fp; fn tp]:")
print(cm)
```
